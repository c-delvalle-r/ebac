# Visualización de Datos: Parte 2: Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
# Comparativa de histogramas entre pandas y seaborn

# df['math score'].plot.hist()
sns.displot(df['math score'], kde=True, bins=10)
```

```python
# Diagramas de dispersión
# Grafico de regresión
sns.regplot(x=df['reading score'], y=df['writing score'])
```

```python
# otra manera de hacer el plot
sns.lmplot(x='reading score', y='writing score', data=df)
```

```python
# Gráfico de dispersión por categoría de genero
sns.lmplot(x='reading score', y='writing score', data=df, hue='race/ethnicity')
```

```python
sns.lmplot(x='reading score', y='writing score', data=df, col='race/ethnicity')
```

```python
sns.lmplot(x='reading score', y='writing score', data=df, col='race/ethnicity')
```

### Personalización de los gráficos de Seaborn

```python
for style in ['dark', 'whitegrid', 'darkgrid', 'ticks']:
    sns.set_style(style=style)
    sns.displot(df['reading score'])
```

```python
from matplotlib.pyplot import title, xlim


sns.set_style('ticks')
figure = sns.displot(df['reading score'], kde=True)
sns.despine(left=True, bottom=True)             # Remoción de ejes por orientación
figure.fig.suptitle('Histogram')
```

```python
# Utilización de matplotlig y seaborn en conjunto

fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True)
sns.distplot(df['writing score'], ax=ax0)
sns.distplot(df.query('gender == "male"')['writing score'], ax=ax1)
ax0.set(xlabel='Puntaje general de escritura')
ax1.set(xlabel = 'Puntaje de escritura para hombres')
ax1.axvline(x=75, label='Objetivo', linestyle='--')
ax1.legend()
```

```python
# Color Palettes

paleta = ['dark', 'pastel', 'colorblind', 'deep', 'muted', 'bright']

for palette in paleta:
    sns.set_palette(palette)
    sns.displot(df['writing score'], kde=True)
```

## Gráficos categóricos

- Stripplotes
- Swarmplots
- Boxplots
- Violin Plots
- Barplots
- Pointplots
- Countplots

```python
# Gráfico de franjas o bandas (Stripplot)

fig, ax = plt.subplots()
sns.stripplot(data=df, y='writing score', x='math score', jitter=True)
ax.axes.xaxis.set_ticks([])
plt.show()
```

```python
sns.stripplot(data=df, x='race/ethnicity', y='writing score', jitter=True)
```

```python
# Diagrama de enjambre (Swarmplot)
# Sirve para evitar empalmes de puntos dentro de la visualización
# Cuando se empalmarían demasiados puntos puede que no todos los puntos vengan en la visualización, eso es lo que marca el "error"

sns.swarmplot(data=df, x='race/ethnicity', y='writing score')
```

```python
# Diagrama de caja (boxplot)
sns.boxplot(data=df, x='race/ethnicity', y='writing score')
```

```python
# Gráfico de violín (violinplot)
# Es una box plot con una función de densidad de Kernel

sns.violinplot(data=df, x='race/ethnicity' , y='writing score')
```

```python
# lvplot (letter-value plot)

sns.catplot(data=df, x='race/ethnicity' , y='writing score', kind='boxen')
```

```python
# bar plot

sns.barplot(data=df, x='race/ethnicity' , y='writing score')
```

```python
# gráfico de barras agrupado por otra variable
sns.barplot(data=df, x='race/ethnicity' , y='writing score', hue='lunch')
```

```python
sns.pointplot(data=df, x='race/ethnicity' , y='writing score', hue='lunch')
```

```python
# countplot

sns.countplot(data=df, y='race/ethnicity', hue='lunch')
```

```python
sns.regplot(data=df, x='reading score' , y='writing score', marker='+')
```

```python
sns.residplot(data=df, x='reading score' , y='writing score')
```

```python
import numpy as np
sns.regplot(data=df, y='writing score', x='reading score', x_estimator=np.mean)
```

```python
# heatmap

sns.heatmap(df.corr(), annot=True)
```

```python
# Tabla cruzada
pd.crosstab(df['gender'], df['race/ethnicity'], values=df['math score'], aggfunc= 'mean')
```

```python
sns.heatmap(pd.crosstab(df['gender'], df['race/ethnicity'], values=df['math score'], aggfunc= 'mean'), annot=True)
```

## Gráficos múltiples

Creación de gráficos múltiples en una sola figura a través de facetgrid, catplot, impolot y jointplot

```python
# FacetGrid

graf = sns.FacetGrid(df, col='race/ethnicity')
graf.map(sns.boxplot, 'math score', order = ['group A', 'group B', 'group C', 'group D', 'group E'])
```

```python
sns.catplot(x='math score' ,data=df, col='race/ethnicity', kind='box')
```

```python
graf = sns.FacetGrid(df, col='race/ethnicity')
graf.map(sns.scatterplot, 'reading score', 'writing score')
```

```python
# Gráficos tabulares de regresión

sns.lmplot(data=df, x='reading score', y='writing score', col='gender', row='race/ethnicity')
```

```python
# gráficos conjuntos

graf = sns.JointGrid(data=df, x='reading score', y='writing score')
graf.plot(sns.regplot, sns.histplot)
```

```python
sns.jointplot(data=df, x='reading score', y='writing score', kind='hex')
```

```python
graf = sns.jointplot(data=df[df['gender'] == 'male'], x='reading score', y='writing score', kind='scatter')
# graf = sns.jointplot(data=df.query('gender == "male"'), x='reading score', y='writing score', kind='scatter')
```
