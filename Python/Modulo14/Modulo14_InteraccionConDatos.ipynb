## Manejo de archivos .csv y .json

```python
# Ejemplo para leer un archivo csv ubicado en otra carpeta con csv reader
import os
import csv
import numpy as np

wd = 'd:/Documentos/DataAnalysis/EBAC/Python/Modulo12'

os.chdir(wd)

with open(r'./netflix_titles.csv', encoding='UTF8') as f:
    csvreader = csv.reader(f)

    header = []
    header = next(csvreader)
    print(header)

    rows = []

    for row in csvreader:
        rows.append(row)
    
    movies = np.array(rows)

print(movies[:2])

# O se puede abrir y cargar directamente a un df de pandas...
```

```python
# Escribir data a un csv usando csv writter

# Cambiar directorio para guardar el archivo en la carpeta del módulo
os.chdir('d:/Documentos/DataAnalysis/EBAC/Python/Modulo14')

header = ['Id', 'Nombre', 'Calificación', 'Género' ]
data = [
    ['1', 'Valorant', 5, 'Online - FPS'],
    ['2', 'Overwatch 2', 4, 'Online - FPS'],
    ['3', 'Skyrim', 5, 'Open World - Medieval - Fantasy'],
]

with open('juegos.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
 
```
