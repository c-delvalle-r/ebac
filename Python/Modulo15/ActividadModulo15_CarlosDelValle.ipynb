### 1. Importación de un sitio web con requests que impresión del contenido

```python
import requests

url = 'https://listado.mercadolibre.com.mx/rtx-3060ti#D[A:rtx%203060ti]'

headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5-0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
}

page = requests.get(url, headers)

print(page.status_code)
print(page.text)
```

### 2. Uso de BeautifulSoup y requests. Parseo de información de un sitio web (como mercadolibre, despegar.com, etc.)

```python
# Utilizando el mismo url del ejercicio pasado

from bs4 import BeautifulSoup

html = BeautifulSoup(page.content, 'html.parser')

product_names = []
product_prices = []

for card in html.select('div.ui-search-result__content-wrapper'):
    product_names += [card.select('.ui-search-item__title')[0].get_text()]
    product_prices+= [card.select('.price-tag-fraction')[0].get_text()]

# Imprimier el listado ordenado de los productos
for i in range(len(product_names)):
    print(f'{product_names[i]} - {product_prices[i]}')
```

### 3. Uso de librería requests con respuestas de API
Se puede ver un listado de APIs gratuitas sin necesidad de pago o credenciales en: [https://mixedanalytics.com/blog/list-actually-free-open-no-auth-needed-apis/](https://mixedanalytics.com/blog/list-actually-free-open-no-auth-needed-apis/)

```python
if requests.get('http://www.boredapi.com/api/activity/').status_code == 200:
    api_response = requests.get('http://www.boredapi.com/api/activity/')
    json_response = api_response.json()
    # print(json_response)

    possible_types = {"education": 'Educación',
                        "recreational": 'Recreación',
                        "social": 'Social',
                        "diy": 'Manualidades',
                        "charity": 'Caridad',
                        "cooking": 'Cocina',
                        "relaxation": 'Relajación',
                        "music": 'Música',
                        "busywork": 'Entretenimiento'}

    # Format API response
    price = 'Gratis' if json_response['price'] == 0 else 'Costo bajo' if json_response['price'] <= .33 else 'Costo medio' if json_response['price'] <= .66 else 'Costo alto'
    link = f'Puedes conocer más en: {json_response["link"]}' if json_response['link'] != '' else ''
    accesibility = 'Dificultad baja' if json_response['accessibility'] <= .33 else 'Dificultad media' if json_response['accessibility'] <= .66 else 'Dificultad alta'

    print(f'Tipo y sugerencia de actividad al azar: {possible_types[json_response["type"]]} - {json_response["activity"]}')
    print(f'Participantes: {json_response["participants"]}')
    print('Costo:', price)
    print('Dificultad:', accesibility)
    print(link)

else:
    print('Error connecting to api')
 
```
