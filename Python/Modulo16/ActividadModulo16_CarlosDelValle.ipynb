```python
# Importar librerías y archivo
import pandas as pd

df = pd.read_csv('d:/Documentos/DataAnalysis/EBAC/Python/Modulo14/fifa_eda.csv')
# 1.1 - Primeras 20 filas del df
df.head(20)
```

```python
# 1.2 - Últimas 5 filas del df
df.tail(5)
```

```python
# 1.3 - Sample de 10 filas del df
df.sample(10)
```

```python
# 2.1 Estadísticas de describe()
df.describe()
```

```python
# 2.2 Tipo de datos del df
df.info()

# A todos los campos numéricos del df les fue asignado el tipo correcto de valor
```

```python
# 4 - Añadir columna Years Playing
df['Years Playing'] = 2022 - df['Joined']
df
```

```python
# 5 - Mostrar a todos los jugadores de México
df.loc[df['Nationality'] == 'Mexico']
```

```python
# 6 - Ordernar datos por la columna 'Release Clause'
df.sort_values(by='Release Clause')
```

```python
# 7 - Generar un dataset que contenga el año (joined) y número de jugadores
df_new = df.groupby(by='Joined').size().rename('Players').reset_index()
df_new
```

```python
# 8 - Generar un gráfico que contenga el año y número de jugadores
df_new.plot.bar(x='Joined', y='Players')
```
