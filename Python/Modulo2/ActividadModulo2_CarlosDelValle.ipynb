1. Genera un archivo de texto a partir de la palabra ”Desperté” y terminando en la
palabra “FIN”, tal y como se ejemplificó en el Módulo 2. Llámalo “ElRamoAzul.txt”
2. Genera una lista que muestre el número de veces que se repite cada palabra
en el texto. Dicha lista deberá ser ordenada de forma alfabética y mostrada a
través de tuplas.
3. Imprime el número de veces que se repite la palabra “ojos”

```python
# 2.
# Abrir archivo de texto y cargar las palabras a una lista
words = []
with open('ElRamoAzul.txt', encoding='utf8') as f:
    for line in f:
        words = words + line.split()

# Hacer el conteo de las palabras y asignarlas a un diccionario
counter = {}
for word in words:
    word = word.lower()     #Cambiar la palabra a minúsculas para contabilizarlas igual sin importar si comienzan los enunciados
    for character in ['.', ',', ' ',':', '-', '¿', '?', '¡', '!', '“', '–']:  # Quitar carácters que pueden ir pegados a las palabras para poderlas contabilizar como la misma
        word = word.replace(character, '')
    
    # Determinar si la palabra ya existe en el diccionario, agragarla si no existe, aumentar en uno el contabilizador si sí existe
    if word not in counter.keys():
        counter[word] = 1
    else: 
        counter[word] += 1


# Ordenar lista y pasarlo a tuplas
tuple_counter = sorted(counter.items())

# Imprimir contador en tuplas
tuple_counter
```

```python
# 4. Imprimir el número de veces que se repite la palabra "ojos"
print('"Ojos" se repite', counter['ojos'], 'veces.')
```
